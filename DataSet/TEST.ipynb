```python
import pandas as pd
```

```python
df = pd.read_csv('D:\Project\Earthquake\DataSet\DATASET_UPDATE.csv')
df['NewTime'] = df['time']
df['time'] = pd.to_datetime(df['time']).dt.strftime("%Y-%m-%d")
df = df[df['time'] >= '2000-01-01']
```

```python
asteroid = pd.read_csv('D:\Project\Earthquake\ASTroidCTrain.csv')
asteroid
```

```python
merged_df = pd.merge(df, asteroid, how='left', left_on='time', right_on='Time', suffixes=('_df', '_asteroid'))
merged_df
```

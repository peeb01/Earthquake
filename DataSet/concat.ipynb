```python
import pandas as pd

df = pd.read_csv('D:\Project\Earthquake\DataSet\query.csv')
df = df[['time', 'latitude', 'longitude', 'mag', 'depth', 'nst', 'gap', 'dmin']].values
dt = pd.read_csv('D:\Project\Earthquake\DataSet\query Last raw.csv')
dt = dt[['time', 'latitude', 'longitude', 'mag', 'depth', 'nst', 'gap', 'dmin']].values
```

```python
_data = []
for i in range(len(dt)):
    _data.append(dt[i])

for j in range(len(df)):
    _data.append(df[j])

column = ['time', 'latitude', 'longitude', 'mag', 'depth', 'nst', 'gap', 'dmin']
data = pd.DataFrame(_data, columns = column)
data.to_csv('datas.csv')
```
